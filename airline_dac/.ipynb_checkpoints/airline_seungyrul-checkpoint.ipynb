```python
numeric_feature = ['Age', 'Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']
ordinal_feature = ['Seat comfort', 'Departure/Arrival time convenient', 'Food and drink', 'Gate location', 
                   'Inflight wifi service', 'Inflight entertainment', 'Online support', 'Ease of Online booking', 
                   'On-board service', 'Leg room service', 'Baggage handling', 'Checkin service', 'Cleanliness', 'Online boarding']
categorical_feature = ['Gender', 'Customer Type', 'Type of Travel', 'Class']

fig, ax = plt.subplots(5, 4, figsize = (10, 15))

plt.suptitle("feature / target", fontsize=10)

feature = ordinal_feature + categorical_feature

xvalue = 0
yvalue = 0

for xstr in feature:
    sns.barplot(x = xstr, y = 'target', data = air_df, ax = ax[yvalue][xvalue])
        
    if xvalue == 3:
        xvalue = 0
        yvalue += 1
    else:
        xvalue += 1
```

```python
air_df_new['Departure Delay in Minutes'] = np.log1p(air_df_new['Departure Delay in Minutes'])
air_df_new['Arrival Delay in Minutes'] = np.log1p(air_df_new['Arrival Delay in Minutes'])

test['Departure Delay in Minutes'] = np.log1p(test['Departure Delay in Minutes'])
test['Arrival Delay in Minutes'] = np.log1p(test['Arrival Delay in Minutes'])
```

```python
air_df_new['Gender'] = air_df_new['Gender'].astype('category').cat.codes
air_df_new['Customer Type'] = air_df_new['Customer Type'].astype('category').cat.codes
air_df_new['Type of Travel'] = air_df_new['Type of Travel'].astype('category').cat.codes
air_df_new['Class'] = air_df_new['Class'].astype('category').cat.codes

test['Gender'] = test['Gender'].astype('category').cat.codes
test['Customer Type'] = test['Customer Type'].astype('category').cat.codes
test['Type of Travel'] = test['Type of Travel'].astype('category').cat.codes
test['Class'] = test['Class'].astype('category').cat.codes

air_df_drop = air_df_new.drop(['id', 'Departure Delay in Minutes', 'Gate location', 'Departure/Arrival time convenient'], axis=1)
test = test.drop(['id', 'Departure Delay in Minutes', 'Gate location', 'Departure/Arrival time convenient'], axis=1)
```

```python
from pycaret.classification import *
clf1 = setup(data = air_df_drop,
             silent = True,
             target = 'target')
```

```python
best = compare_models(n_select=3, fold = 10)
```

```python
blend_3_soft = blend_models(estimator_list=best, method='soft', choose_better = True)
predict_model(blend_3_soft).head(10)
```

```python
final_blend_3_soft = finalize_model(blend_3_soft)
pred_test = predict_model(final_blend_3_soft, data=test)
```
